```python
def change_value(df, name, feature, new_value):
    df.loc[df[name] == name, feature] = new_value
    return df
```

```python
change = input('Do you want to change any values in the QB dataframes? ')
while change == 'y':
    site = input('Which site? ')
    name = input('Which player? ')
    feature = input('Which feature? ')
    new_value = float(input('Enter new value '))
    if site == 'FD':
        fd_qb_df = change_value(fd_qb_df, name, feature, new_value)
    else:
        dk_qb_df = change_value(fd_qb_df, name, feature, new_value)
    change = ('Do you want to change any other values in the QB dataframes? ')
```

```python
change = input('Do you want to change any values in the FLEX dataframes? ')
while change == 'y':
    site = input('Which site? ')
    name = input('Which player? ')
    feature = input('Which feature? ')
    new_value = float(input('Enter new value '))
    if site == 'FD':
        fd_flex_df = change_value(fd_flex_df, name, feature, new_value)
    else:
        dk_flex_df = change_value(fd_flex_df, name, feature, new_value)
    change = ('Do you want to change any other values in the QB dataframes? ')
```
