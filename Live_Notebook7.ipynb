```python
#Ask user if they want to alter data to reflect a starting QB change
qb_update = input('Are there any starting QB changes? (y or n) ').lower()

# If there are QB changes
while qb_update == 'y':
    fd_list = input('Is this QB in the FanDuel list? (y or n) ').lower()
    
    # If QB is in the FanDuel list
    if fd_list == 'y':
        name = input('Enter name of new starting quarterback: ')
        fd_qb_df.loc[fd_qb_df['name'] == name, 'QB_role'] = 1
        backup = input('Enter name of previous starting quarterback: ')
        fd_qb_df.loc[fd_qb_df['name'] == backup, 'QB_role'] = 2
    
    dk_list = input('Is this QB in the DraftKings list? (y or n) ')
    
    # If QB is in the DraftKings list
    if dk_list == 'y':
        name = input('Enter name of new starting quarterback: ')
        dk_qb_df.loc[dk_qb_df['name'] == name, 'QB_role'] = 1
        backup = input('Enter name of previous starting quarterback: ')
        dk_qb_df.loc[dk_qb_df['name'] == backup, 'QB_role'] = 2
        
    qb_update = input('Are there any other starting QB changes? (y or n) ').lower()


# Dictionary and list for column renaming.
# Because we have variables in the flex dataset that reflect the starting QB, flex data changes in QB_role changes
qb_column_dict = {
    'completions_L8': 'qb_comp', 'attempts_L8': 'qb_att', 'passing_yards_L8': 'qb_yds',
    'passing_tds_L8': 'qb_pass_td', 'interceptions_L8': 'qb_int', 'comp_pct': 'qb_comp_pct',
    'yds_per_attempt': 'qb_yds_per_att', 'td_pct': 'qb_td_pct', 'int_pct': 'qb_int_pct'
}

qb_column_list = list(qb_column_dict.values())

#temp df of starting QBs only
temp_fd_qb_df = fd_qb_df[fd_qb_df['QB_role'] == 1]
temp_fd_qb_df = temp_fd_qb_df.rename(columns=qb_column_dict)
temp_fd_qb_df = temp_fd_qb_df[['season', 'week', 'team', 'opponent'] + qb_column_list]

#Dropping old QB-related data from flex data, and replacing with new data
fd_flex_df.drop(columns=qb_column_list, inplace=True)
fd_flex_df = pd.merge(fd_flex_df, temp_fd_qb_df, on=['season', 'week', 'team', 'opponent'], how='left')
        
#         # Drop old columns from fd_flex_df
#         fd_flex_df.drop(columns=qb_column_list, inplace=True)
        
#         # Merge the dataframes on common columns
#         fd_flex_df = pd.merge(fd_flex_df, temp_fd_qb_df, on=['season', 'week', 'team', 'opponent'], how='left')
# except NameError:
#     print('fd_list is not defined.')

temp_dk_qb_df = dk_qb_df[dk_qb_df['QB_role'] == 1]
temp_dk_qb_df = temp_dk_qb_df.rename(columns=qb_column_dict)
temp_dk_qb_df = temp_dk_qb_df[['season', 'week', 'team', 'opponent'] + qb_column_list]

dk_flex_df.drop(columns=qb_column_list, inplace=True)
dk_flex_df = pd.merge(dk_flex_df, temp_dk_qb_df, on=['season', 'week', 'team', 'opponent'], how='left')
# try:
#     if dk_list == 'y' | dk_list == None:
#         # Filter for starting QBs
#         temp_dk_qb_df = dk_qb_df[dk_qb_df['QB_role'] == 1]
        
#         # Rename columns
#         temp_dk_qb_df = temp_dk_qb_df.rename(columns=qb_column_dict)
        
#         # Select relevant columns
#         temp_dk_qb_df = temp_dk_qb_df[['season', 'week', 'team', 'opponent'] + qb_column_list]
        
#         # Drop old columns from fd_flex_df
#         dk_flex_df.drop(columns=qb_column_list, inplace=True)
        
#         # Merge the dataframes on common columns
#         dk_flex_df = pd.merge(dk_flex_df, temp_dk_qb_df, on=['season', 'week', 'team', 'opponent'], how='left')
# except NameError:
#     print('dk_list is not defined.')

#Dropping any QBs who are on IR out listed as Out
#fd_qb_df = fd_qb_df.reset_index()
fd_qb_df = fd_qb_df[~(fd_qb_df['status'].isin(['IR', 'O']))]

fd_qb_df = fd_qb_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])
fd_flex_df = fd_flex_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])

dk_qb_df = dk_qb_df[~(dk_qb_df['status'].isin(['IR', 'O']))]

dk_qb_df = dk_qb_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])
dk_flex_df = dk_flex_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])

```
