```python
def change_value(df, name, feature, new_value):
    df.loc[df[name] == name, feature] = new_value
    return df
```

```python
change_def = input('Do you want to change any values in the def dataframes? ')
while change_def == 'y':
    site = input('Which site? ')
    name = input('Which player? ')
    feature = input('Which feature? ')
    new_value = float(input('Enter new value '))
    if site == 'FD':
        fd_def_df = change_value(fd_def_df, name, feature, new_value)
    else:
        dk_def_df = change_value(dk_def_df, name, feature, new_value)
    change_def = ('Do you want to change any other values in the QB dataframes? ')
    
change_qb = input('Do you want to change any values in the QB dataframes? ')
while change_qb == 'y':
    site = input('Which site? ')
    name = input('Which player? ')
    feature = input('Which feature? ')
    new_value = float(input('Enter new value '))
    if site == 'FD':
        fd_qb_df = change_value(fd_qb_df, name, feature, new_value)
    else:
        dk_qb_df = change_value(dk_qb_df, name, feature, new_value)
    change_qb = ('Do you want to change any other values in the QB dataframes? ')

change_flex = input('Do you want to change any values in the FLEX dataframes? ')
while change_flex == 'y':
    site = input('Which site? ')
    name = input('Which player? ')
    feature = input('Which feature? ')
    new_value = float(input('Enter new value '))
    if site == 'FD':
        fd_flex_df = change_value(fd_flex_df, name, feature, new_value)
    else:
        dk_flex_df = change_value(dk_flex_df, name, feature, new_value)
    change_flex = ('Do you want to change any other values in the QB dataframes? ')

fd_def_cols = list(fd_def_df.columns)
dk_def_cols = list(dk_def_df.columns)
fd_qb_cols = list(fd_qb_df.columns)
dk_qb_cols = list(dk_qb_df.columns)
fd_flex_cols = list(fd_flex_df.columns)
dk_flex_cols = list(dk_flex_df.columns)

def_x_cols = ['spread_line',
 'home_team',
 'outdoors',
 'grass',
 'total_line',
 'opp_total',
 'opp_vs_def_L8',
 'opp_vs_def_L8_lin',
 'opp_vs_def_L8_exp']

qb_x_cols = ['FD_Pts_DvP', 'DK_Pts_DvP', 'QB_role', 'completions_L8', 'attempts_L8',
       'passing_yards_L8', 'passing_tds_L8', 'interceptions_L8', 'sacks_L8',
       'passing_air_yards_L8', 'pacr_L8', 'carries_L8', 'rushing_yards_L8',
       'rushing_tds_L8', 'comp_pct', 'yds_per_attempt', 'td_pct', 'sack_pct',
       'int_pct', 'wind', 'div_game', 'spread_line', 'total_line', 'outdoors',
       'grass', 'home_team', 'pred_total', 'opp_total']

flex_x_cols = ['DK_Pts_RB_DvP', 'DK_Pts_TE_DvP', 'DK_Pts_WR_DvP', 'FD_Pts_RB_DvP',
       'FD_Pts_TE_DvP', 'FD_Pts_WR_DvP', 'rushing_yards_L8', 'rushing_tds_L8',
       'rushing_fumbles_lost_L8', 'receptions_L8', 'receiving_yards_L8',
       'receiving_tds_L8', 'receiving_fumbles_lost_L8', 'targets_L8',
       'carries_L8', 'receiving_yards_after_catch_L8',
       'receiving_air_yards_L8', 'target_share_L8', 'carry_share_L8',
       'in_10_share_L8', 'yards_per_carry_L8', 'yards_per_reception_L8',
       'yards_per_target_L8', 'wind', 'div_game', 'spread_line', 'total_line',
       'outdoors', 'grass', 'home_team', 'pred_total', 'opp_total', 'qb_comp',
       'qb_att', 'qb_yds', 'qb_pass_td', 'qb_int', 'qb_comp_pct',
       'qb_yds_per_att', 'qb_td_pct', 'qb_int_pct', 'pos_RB', 'pos_TE',
       'pos_WR']

# List of dataframes
df_list = {
    'fd_def_df': fd_def_df,
    'fd_qb_df': fd_qb_df,
    'fd_flex_df': fd_flex_df,
    'dk_def_df': dk_def_df,
    'dk_qb_df': dk_qb_df,
    'dk_flex_df': dk_flex_df
}

# List to store whether each dataframe has missing values
missing_bool_list = []

# Check for missing values in each dataframe and store in the list
# for df in df_list:
#     missing_values = df.isnull().any().any()  # Check for any missing values
#     missing_bool_list.append(missing_values)

for df_name, df in df_list.items():
    missing_values = df.isnull().any().any()  # Check for any missing values
    missing_bool_list.append((df_name, missing_values))

# Create a dictionary mapping each dataframe to its missing value status
missing_dict = dict(missing_bool_list)

# Iterate over the dictionary to handle dataframes with missing values
# Iterate over the dictionary to handle dataframes with missing values
for df_name, has_missing in missing_dict.items():
    if has_missing:
        print(f"{df_name} has missing values")

        # Specify which columns have missing values
        missing_columns = df_list[df_name].columns[df_list[df_name].isna().any()].tolist()
        print(f"Columns with missing values in {df_name}: {missing_columns}")

        # Display rows with missing values
        display(df_list[df_name][df_list[df_name].isna().any(axis=1)])

        # Allow the user to fill missing values
        fill = input('Do you want to fill a value? ').lower()
        while fill == 'y':
            col = input(f"Which column do you want to fill? Choose from {missing_columns}: ")
            name = input('Which player? ')
            new_value = float(input('Enter new value: '))
    
            # Update the dataframe where the 'name' column matches the input player name
            df_list[df_name].loc[df_list[df_name]['name'] == name, col] = new_value
            fill = input('Do you want to fill another value? (y/n) ').lower()

fd_def_df = fd_def_df.dropna()
fd_qb_df = fd_qb_df.dropna()
fd_flex_df = fd_flex_df.dropna()
dk_def_df = dk_def_df.dropna()
dk_qb_df = dk_qb_df.dropna()
dk_flex_df = dk_flex_df.dropna()

fd_def_df = fd_def_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])
dk_def_df = dk_def_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])
fd_qb_df = fd_qb_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])
dk_qb_df = dk_qb_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])
fd_flex_df = fd_flex_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])
dk_flex_df = dk_flex_df.set_index(['name', 'team', 'position', 'salary', 'opponent', 'status', 'season', 'week'])

fd_flex_df = fd_flex_df[flex_x_cols]
dk_flex_df = dk_flex_df[flex_x_cols]
fd_qb_df = fd_qb_df[qb_x_cols]
dk_qb_df = dk_qb_df[qb_x_cols]
fd_def_df = fd_def_df[def_x_cols]
dk_def_df = dk_def_df[def_x_cols]

all_conditions = all([
    list(fd_def_df.columns) == def_x_cols,
    list(dk_def_df.columns) == def_x_cols,
    list(fd_qb_df.columns) == qb_x_cols,
    list(dk_qb_df.columns) == qb_x_cols,
    list(fd_flex_df.columns) == flex_x_cols,
    list(dk_flex_df.columns) == flex_x_cols
])

if not all_conditions:
    print("Not all column orders match training data.")

change_datasets = input('Did dataset values change? ')
if change_datasets == 'y':
    fd_flex_df.to_csv('FD_FLEX_for_model_' + str(season) + '_' + str(week) + '.csv')
    dk_flex_df.to_csv('DK_FLEX_for_model_' + str(season) + '_' + str(week) + '.csv')
    fd_qb_df.to_csv('FD_qb_for_model_' + str(season) + '_' + str(week) + '.csv')
    dk_qb_df.to_csv('DK_qb_for_model_' + str(season) + '_' + str(week) + '.csv')
    fd_def_df.to_csv('FD_def_for_model_' + str(season) + '_' + str(week) + '.csv')
    dk_def_df.to_csv('DK_def_for_model_' + str(season) + '_' + str(week) + '.csv')
```
